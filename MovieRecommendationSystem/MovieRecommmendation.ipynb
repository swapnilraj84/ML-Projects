```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['genres' , 'keywords' , 'tagline' , 'cast' , 'director']
```

```python
print(selected_features)
```

```python
for feature  in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
list_all = movies_data['title'].tolist()
```

```python
print((list_all))
```

```python
movie_name = input('Enter Movie Name')
```

```python
find_close_match = difflib.get_close_matches(movie_name , list_all);
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_movie = movies_data[movies_data.title == close_match]['index'].values[0];
print(index_movie)
```

```python
similarity_score = list(enumerate(similarity[index_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for You')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title = movies_data[movies_data.index == index]['title'].values[0]
    if(i<30):
        print(i , '.',title)
        i+=1
```
